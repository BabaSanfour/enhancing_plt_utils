```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))  

x_after = np.linspace(0, 2 * np.pi, 100)
y1_after = np.sin(x_after)
y2_after = np.cos(x_after)
y3_after = np.tanh(x_after)

plt.plot(x_after, y1_after, label='Sin(x)', linestyle='-',)
plt.plot(x_after, y2_after, label='Cos(x)', linestyle='--',)
plt.plot(x_after, y3_after, label='Tanh(x)', linestyle='-.',)
plt.yticks([-1, -0.5, 0, 0.5, 1]) 
plt.title('Before')  
plt.xlabel('X-axis')  
plt.ylabel('Y-axis')  


plt.annotate('Maximum', xy=(np.pi/2, 1), xytext=(np.pi/6, 1.3),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=12, color='black')
plt.legend()
plt.show()
```

```python
def get_color_matrix(list_colors):
    color_matrix = np.zeros((2, len(list_colors), 4))

    for color_idx, color in enumerate(list_colors):
        rgba_color = plt.cm.colors.to_rgba(color)
        color_matrix[:, color_idx, :4] = rgba_color
    return color_matrix    
```

```python
plt.figure(figsize=(18, 6))
cmap = get_color_matrix(get_color_map('one_piece'))
plt.figure(figsize=(18, 6))
plt.imshow(cmap, aspect='auto', extent=[0, len(cmap), 0, 2])
plt.title('One piece Map')
plt.yticks([])  
plt.xticks([])  
plt.show()
```

```python
# What about continious color maps?
from color_maps import continuous_color_map
```

```python
# from -1 t0 1:
c_cmap = continuous_color_map('pink_ocean')
test_values = np.linspace(-1, 1, 100)
colors = c_cmap.to_rgba(test_values)  
fig, ax = plt.subplots(figsize=(18, 4))
sc = ax.scatter(test_values, np.zeros_like(test_values), c=colors, marker='|', s=1000)

# Use c_cmap directly for the colorbar
cbar = plt.colorbar(c_cmap, ax=ax, orientation='vertical', ticks=[-1, 0, 1])
cbar.set_label('Color Mapping')
cbar.set_ticklabels(['-1', '0', '1'])

plt.title('Continuous Colormap for Values from -1 to 1')
plt.xlabel('Values')
plt.yticks([])  
plt.show()
```
